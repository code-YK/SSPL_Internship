In-built methods to access attributes(kinda a variables)
1. getattr(obj, attr, default): Retrieves an attribute's value; returns default if missing.
2. hasattr(obj, attr): Checks if an attribute exists, returning True or False.
3. setattr(obj, attr, value): Assigns a value to an attribute, creating it if absent.
4. delattr(obj, attr): Deletes an attribute; accessing it afterward raises an error.

```python
class Employee:
    name='kuldeep'
    age='20'
    post='Intern'
```

```python
ep=Employee()
print(getattr(ep,'name'))
print(hasattr(ep,'age'))
setattr(ep,'feild','Aiml')
print(getattr(ep,'feild'))
```

```python
delattr(ep,'feild')
print(hasattr(ep,'feild'))
```

Create s student class with constructor that takes name and marks of three subjects and create a method to print its average.

```python
class StudentMarks:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def avg(self):
        sum=0
        for i in self.marks:
            sum+=i
        print(f"{self.name}'s average is {sum/3}")
```

static methods don't have self as first parameters, they are decorated by @staticmethod.

```python
class Greet:
    @staticmethod
    def hello():
        print("Hello it is a static method")
    #static method with parameter
    @staticmethod
    def hii(a):
        print("hii "*a)
g=Greet()
g.hello()
g.hii(5)
```

class methods are the methods bounded to the class and can only access the class level attributes, @classmethod decorater is required

```python
class Humans:
    ex='Running'
    @classmethod
    def exercise(cls):
        print(f"Human is {cls.ex}")

h=Humans()
h.exercise()
```

Problem Statement: Create a class Point for a 2D coordinate system. Its __init__ method should take x and y coordinates. Add a method distance_from_origin to calculate the Euclidean distance from (0,0). Create a Point at (3, 4) and print its distance.

```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def euclid_dis(self):
        distance= (self.x**2 + self.y**2)**0.5
        return distance
```

```python
p=Point(3,4)
print(f"Distance between {p.x,p.y} and origin (0,0) is {p.euclid_dis()}")
```

```python
class Line:
    def __init__(self,cordi1,cordi2):
        self.x1,self.y1=cordi1
        self.x2,self.y2=cordi2

    def distance(self):
        dis_x=self.x1 - self.x2
        dis_y=self.y1 - self.y2
        dis=(dis_x**2 + dis_y**2)**0.5
        print(f"Length of line is {dis} units")

    def slope(self):
        dis_x=self.x1 - self.x2
        dis_y=self.y1 - self.y2
        sl=dis_y/dis_x
        print(f"SLope of line is {sl}")
```

```python
l=Line((1,1),(2,2))
l.distance()
l.slope()
```

Real time bank transaction

```python
class Bank:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f'''{amount} deposited successfully!!''')
        print(f'''Current Balance : $ {self.balance}''')

    def withdraw(self,amount):
        if self.balance >= amount:
            self.balance-=amount
            print(f'''{amount} withdrawn successfully!!''')
            print(f'''Current Balance : $ {self.balance}''')
        else:
            print('Insufficient balance!!')

    def checkBal(self):
        print(f'''Current Balance : $ {self.balance}''')
```
