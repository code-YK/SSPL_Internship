List Comprehension
1. Write a program to create a list of squares of all even numbers from 1 to 20 using list comprehension.

```python
lst=[x for x in range(21)]
print(lst)
sq_lst=[x**2 for x in range(21) if x%2==0]
print(sq_lst)
```

2. Write a program to extract all vowels from a given string into a list using list comprehension.



```python
strs=input("Enter a string:").lower()
vowels="aeiou"
lst=[x for x in strs if x in vowels]
print(lst)
```

3. Write a program to convert a list of temperatures in Celsius to Fahrenheit using list comprehension and lambda function.

```python
c_lst=[37,25,100]
# celsis to fahrenheit : c*9/5 + 32
f_lst=[(lambda x: (x*9/5)+32)(temp) for temp in c_lst]
for i in range(len(c_lst)):
    print(f"{c_lst[i]}`C = {f_lst[i]}`F")
```

Set Comprehension
1. Write a program to generate a set of unique characters from a given string using set comprehension.

```python
strs=input("Enter a string:").lower()
lst=[ch for ch in strs]
unique_ch=set(lst)
print(unique_ch)
```

2. Write a program to find all numbers from 1 to 50 that are divisible by both 3 and 5 using set comprehension.

```python
num_set={x for x in range(1,50,1) if x%15==0}
print(num_set)
```

Dictionary Comprehension
1. Write a program to create a dictionary where keys are numbers from 1 to 5 and values are their squares using dictionary comprehension.

```python
d={num:num**2 for num in range(1,6)}
print(d)
```

2. Given a list of words, write a program to create a dictionary with words as keys and their lengths as values using dictionary comprehension.

```python
lst=['hi','hello','good morning']
l_dict = {words:len(words) for words in lst}
print(l_dict)
```

3. Given two lists: keys = ['a', 'b', 'c'] and values = [1, 2, 3], create a dictionary mapping each key to its corresponding value using dictionary comprehension.

```python
keys = ['a', 'b', 'c']
values = [1, 2, 3]
# using dictionary comprehension using zip
d={k:v for k,v in zip(keys,values)}
print(d)
```

Tupple related problems
1. Write a program to create a tuple of even numbers from 1 to 20 using a generator expression.

```python
tup=tuple(x for x in range(1,20) if x%2==0)
print(tup)
```

2. Write a program to create a tuple of the first letters of each word in a list using comprehension.

```python
lst=['ram','krishna','shiv']
f_tup=tuple(words[0] for words in lst)
print(f_tup)
```

```python
n= input("Enter a number:")
if not type(n) is int:
    raise TypeError("Enter a valid integer")
```

```python
def division(a,b):
    try:
        result=a/b
    except ZeroDivisionError as e:
        print(e)
    except ValueError:
        print("please enter valid input")
    finally:
        print("program executed successfully!")
    return result
```

```python
m=int(input("enter a number:"))
n=int(input("enter a number:"))
print(f"{m} divided by {n} gives {division(m,n)}")
```
