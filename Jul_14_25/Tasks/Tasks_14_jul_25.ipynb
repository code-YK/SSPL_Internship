1. Base Class: Employee

    Attributes: name, employee_id

    Method: get_details() â†’ prints basic info

    Subclass: Developer

    Extra Attribute: programming_language

    Override get_details() to include language info

    Subclass: Manager

    Extra Attribute: team_size

    Override get_details() to include team size

    Create a list of various Employee, Developer, and Manager objects.

    Use a loop to call get_details() for each object (to show polymorphism).



```python
class Employee:
    def __init__(self,name,e_id):
        self.name=name
        self.e_id=e_id

    def get_details(self):
        print(f'''
        Name : {self.name}
        Employee id : {self.e_id} ''')
```

```python
class Devloper(Employee):
    def __init__(self,name,e_id,program_lang):
        super().__init__(name,e_id)
        self.program_lang = program_lang

    def get_details(self):
        print(f'''
        Name : {self.name}
        Employee id : {self.e_id}
        Programming Langauge : {self.program_lang}''')
```

```python
class Manager(Employee):
    def __init__(self,name,e_id,team_size):
        super().__init__(name,e_id)
        self.team_size=team_size

    def get_details(self):
        print(f'''
        Name : {self.name}
        Employee id : {self.e_id}
        Team Size : {self.team_size}''')
```

```python
e1 = Employee("Kuldeep", 101)
d1 = Devloper("Vedant", 102, "Python")
m1 = Manager("Rahul", 103, 5)

emp_lst = [e1, d1, m1]
for emp in emp_lst:
    emp.get_details()
```

2. Implement polymorphism with a common method

```python
class Animal:
    def make_sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def make_sound(self):
        print("Dog is barking!!")

class Cat(Animal):
    def make_sound(self):
        print("Meow~ Meow~")

class Lion(Animal):
    def make_sound(self):
        print("Roar!!!!")
```

```python
animal_lst = [Dog(), Cat(), ()]

for animal in animal_lst:
    animal.make_sound()
```
