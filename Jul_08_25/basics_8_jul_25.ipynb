```python
# sum of digits in number (using arithmetic operators)
```

```python
num=int(input('Enter a number:'))
sum=0
d=0
while num!=0:
    d = num % 10
    num=num//10
    sum+=d
print(sum)
```

```python
# sum of digits in number using string
```

```python
num=input('Enter a number:')
num=str(num)
sum=0
for i in num:
    sum+=int(i)
print(sum)
```

```python
#Factorial using recursing function
```

```python
num=int(input("Enter number:"))
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
print(fact(num))
```

```python
# string reversal
```

```python
s=input("Enter a string:")
def reverse_s(Str):
    lst=list(Str)
    rev_lst=[]
    for i in range(len(lst)-1,-1,-1):
        rev_lst.append(lst[i])
    print("".join(rev_lst))
reverse_s(s)
```
