```python
a=int(input("Enter a number:"))
b=int(input("Enter 2nd number:"))
o=input("Enter the operator:")
if o=="+":
    print(f"{a} + {b} = {a+b}")
elif o=="-":
    print(f"{a} - {b}={a-b}")
elif o=="*":
    print(f"{a} * {b}={a*b}")
elif o=="/":
    print(f"{a} / {b}={a/b}")
elif o=="^" or o=="**":
    print(f"{a} ^ {b}={a**b}")
else:
    print("Please enter a valid operator !!")
```

```python
#Count Vowels
```

```python
St=input("Enter any string:").lower()
vowels = "aeiou"
count_v=0
for ch in St:
    if ch in vowels:
        count_v+=1
print(f"Vowels count = {count_v}")
```

```python
marks={'kuldeep':95 ,'vedant':85 ,'shlok':80 }
print(marks)
max_marks=0
rank_1=None
for keys in marks:
    if marks[keys]>max_marks:
        max_marks = marks[keys]
        rank_1=keys
print(rank_1)
```

```python
St=input("Enter any string:").lower()
f={}
for i in St:
    f[i]=f.get(i,0)+1
print(f)
```

```python
# most obtained character in the string
```

```python
St=input("Enter any string:").lower()
f={}
for i in St:
    f[i]=f.get(i,0)+1
max_ch=max(f,key=f.get)
print(max_ch)
```

```python
#sorting list using for loops
```

```python
lst=[10,2,3,34,12]
for i in range(len(lst)):
    min_i = i
    for j in range(i+1,len(lst)):
        if lst[j]<lst[min_i]:
            min_i=j
    lst[i],lst[min_i]=lst[min_i],lst[i]
print(lst)
```

```python
# roman to integer
```

```python
rn=input("Enter roman number:").upper()
roman={'I':1,'V':5,'X':10,'L':50, 'C':100, 'D':500, 'M':1000}
sum=0
pre_v = 0
for ch in rn[::-1]:
    val=roman[ch]
    if val>=pre_v:
        sum+=val
    else:
        sum-=val
    pre_v=val
print(sum)
```
