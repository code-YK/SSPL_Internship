Single thread : all the programs we runned till now works on single thread.


```python
a=10
b=20
print(a+b)
```

the above running cell works on single thread. To understand it more clearly lets go through sleep class of time


```python
import time

def task1():
    time.sleep(2)
    print("Task 1 is executing")

def task2():
    time.sleep(4)
    print("Task 2 is executing")

def task3():
    time.sleep(4)
    print("Task 3 is executing")
```

as the sleep interval is used and all the three methods work on single thread, task1 will execute after 2 sec , task 2 will execute at 6th sec and task3 will be executed at 10th sec

```python
task1()
task2()
task3()
```

To run all the three tasks seperately and at same time, we need to use multithread

```python
import threading
```

```python
import time

def task1():
    print("Task 1 is executing")
    time.sleep(2)
    print("Task 1 is finished after 2s")

def task2():
    print("Task 2 is executing")
    time.sleep(4)
    print("Task 2 is finished after 4s")

def task3():
    print("Task 3 is executing")
    time.sleep(6)
    print("Task 3 is finished after 6s")
```

```python
thread1  = threading.Thread(target=task1)
thread1.start()

thread2  = threading.Thread(target=task2)
thread2.start()

thread3  = threading.Thread(target=task3)
thread3.start()

# Calling thread.join() on a thread will block the main thread until that specific thread finishes.
thread1.join()
thread2.join()
thread3.join()

print("All tasks started running at same time and finished")
```

Passing methods with arguements in thread objects.

note: to pass single arguement we need to put comma at end as args take values in tuple and to create tupple with one element is like (10,)

```python
import threading


def print_cube(num):
    print(f"Cube: {num*num*num}")


def print_square(num):
    print(f"Square: {num * num}")

t1 = threading.Thread(target=print_square, args=(10,))
t2 = threading.Thread(target=print_cube, args=(10,))

t1.start()
t2.start()

t1.join()
t2.join()

print("Done!")
```

###THREADPOOLEXECUTOR
-> when we need to create multiple threads , creating them manually creates a tough task. ThreadPoolExecuter creates them automatically (like looping) and executes them.

-> It is a class of concurrent/futures file

```python
from concurrent.futures import ThreadPoolExecutor
import time

def task(name):
    print(f"{name} is starting")
    time.sleep(2)
    print(f"{name} is done")
with ThreadPoolExecutor(max_workers=3) as executor:
    executor.submit(task, "Thread-1")
    executor.submit(task, "Thread-2")
    executor.submit(task, "Thread-3")
```

```python
from concurrent.futures import ThreadPoolExecutor
import time

def cube(num):
    time.sleep(2)
    return num**3

with ThreadPoolExecutor() as executer:
    cubes = executer.map(cube,range(1,5))
    print(list(cubes))
```
