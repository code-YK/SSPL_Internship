This returns a list of 10 users in JSON format.


```python
import requests
response = requests.get('https://jsonplaceholder.typicode.com/users')
users = response.json()
print(type(users))
for i in users:
    print(f'''
    Name : {i['name']}
    Email : {i['email']}''')
```

```python
for i in users:
    name=i['name'].lower()
    if name[0] == 'c':
        print(f'''
    Name : {i['name']}
    Email : {i['email']}''')
```
