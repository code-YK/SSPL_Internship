#TASK-I: Student Record System Using Classes

```python
class Student:
    def __init__(self,name,roll_no,marks):
        self.name=name
        self.roll_no=roll_no
        self.marks=marks

    def display_info(self):
        print(f'''
        Name : {self.name}
        Roll No. : {self.roll_no}
        Marks : {self.marks}
        Pass : {self.has_passed()}''')
        print('--'*10)

    def has_passed(self):
        return self.marks >= 40.00
```

```python
s1=Student('John',1,67.35)
s2=Student('Rahul',2,90.00)
s3=Student('Om',3,50.00)
s4=Student('Tanya',4,37.06)
s5=Student('Raju',5,00.00)
```

```python
s1.display_info()
s2.display_info()
s3.display_info()
s4.display_info()
s5.display_info()
```

#TASK-II: Vehicle Rental Service System


```python
class Vechile:
    company_name = "QuickRide Rentals"
    def __init__(self,vechile_id:int,types:str,brand:str,rental_rate_per_hour:float,is_available:bool):
        self.vechile_id=vechile_id
        self.brand=brand
        self.rental_rate_per_hour=rental_rate_per_hour
        self.is_available=is_available
        self.types=types

    def display_info(self):
        print('--'*4 + Vechile.company_name + '--'*4)
        print(f'''Vehicle ID : {self.vechile_id}
Type : {self.types}
Brand : {self.brand}
Rate(per hour) : {self.rental_rate_per_hour}
Availabe : {'Yes' if self.is_available else 'No'}''')
        print('--'*12)

    @classmethod
    def change_company_name(cls, new_name):
        cls.company_name=new_name

    @staticmethod
    def calculate_rent(hours:int, rate:float):
        return hours*rate
```

```python
v1=Vechile(1,'Bike','Yamaha','700',True)
v1.display_info()
```

```python
print(v1.calculate_rent(2,700))
```
