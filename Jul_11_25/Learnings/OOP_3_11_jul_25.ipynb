#Inheritance

1. single level

```python
class Base:
    def show(self):
        print("Base class's method")

class Derived(Base):
    def display(self):
        print("This is derived class's method")
```

```python
obj=Derived()
obj.show()
obj.display()
```

2. Multi-level

```python
class Grandparent:
    def g_show(self):
        print("Child using g_show() method")
class Parent(Grandparent):
    def p_show(self):
        print("Child using g_show() method")
class Child(Parent):
    def c_show(self):
        print("I am a child, i have inherited properties of both")
```

```python
obj=Child()
obj.c_show()
obj.p_show()
obj.g_show()
```

3. Multiple inheritance

```python
class Father:
    f_name='Maheshbhai'
    def f_show(self):
        print(f"Father name : {self.f_name}")

class Mother:
    m_name='Shobhaben'
    def m_show(self):
        print(f"Mother name : {self.m_name}")

class Child(Father,Mother):
    def c_show(self):
        print("I have parent properties")
```

```python
obj=Child()
obj.f_show()
obj.m_show()
obj.c_show()
```

#Method overriding

```python
class Car:
    def run(self):
        print("car is running")

class Ferrari(Car):
    def run(self):
        print("Ferrari is running")
```

```python
f=Ferrari()
f.run()
```

```python
class Car:
    def __init__(self):
        self.call="car class is invoked"
        print(self.call)
    def run(self):
        print("car is running")

class Ferrari(Car):
    def __init__(self):
        super().__init__()
        super().run()
    def run(self):
        print("Ferrari is running")
```

```python
f=Ferrari()
f.run()
```

Hierarchical Inheritance

```python
class Animal:
    def sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")
```

```python
a=Animal()
c=Cat()
d=Dog()
a.sound()
c.sound()
d.sound()
```
