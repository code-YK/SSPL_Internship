```python
class Car:
    def __init__(self,model,speed):
        self.model=model
        self.speed=speed
        print("Object is created")
    def accelerate(self,s_increment):
        self.speed+=s_increment
        print(f"Car is accelerated !!")
        print(f"Speed is {self.speed} km/hr ")
    def stop(self):
        self.speed=0
        print(f"{self.model} is Stoppedd!!")
```

```python
c1=Car('Honda civic',110)
c1.accelerate(10)
```

```python
c1.stop()
```

```python
class Student:
    total_students = 0

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        Student.total_students += 1

    def get_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        else:
            return "F"

    @classmethod
    def get_total_students(cls):
        return cls.total_students

    @staticmethod
    def is_valid_marks(marks):
        if marks > 100 or marks < 0:
            return False
        else:
            return True
```

```python
s1=Student('Kuldeep',95)
s2=Student('Vedant',70)
s3=Student('Rahul',50)
print(f"Total students = {Student.get_total_students()}")
print(f"{s1.name} has secured {s1.get_grade()}")
print(f"{s2.name} has secured {s2.get_grade()}")
print(f"{s3.name} has secured {s3.get_grade()}")
print(f"Is 150 marks valid? {Student.is_valid_marks(150)}")
print(f"{s1.name} marks are valid? {s1.is_valid_marks(s1.marks)}")
```
